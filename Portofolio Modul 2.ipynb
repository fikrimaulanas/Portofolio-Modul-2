## Menghitung jumlah outlier per kolom numerik berdasarkan Interquartile Range.

```python
def count_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    return ((series < lower) | (series > upper)).sum()

numeric_cols = df.select_dtypes(include=['int64','float64']).columns
outlier_counts = {col: count_outliers(df[col]) for col in numeric_cols}
outlier_counts
```

```python
# 1. Salin data yang sudah dibersihkan
df_lrfm = df.copy()
```

```python
# 2. Definisikan kolom LRFM
lr_cols = {
    'L': 'Months_on_book',
    'R': 'Months_Inactive_12_mon',
    'F': 'Total_Trans_Ct',
    'M': 'Total_Trans_Amt'
}
```

```python
# 3. Untuk R, invert nilainya agar recency kecil → skor tinggi
df_lrfm['R_inverted'] = df_lrfm[lr_cols['R']].max() - df_lrfm[lr_cols['R']]
```

```python
# 4. Hitung skor 1–5 berdasarkan rank percentile
def score_from_rank(series, n_bins=5):
    # Hitung peringkat dari 1 hingga N
    ranks = series.rank(method='first')
    # Konversi ke persentil (0-1)
    percentiles = (ranks - 1) / (len(series) - 1)
    # Bagi ke dalam bins, lalu +1 agar skornya 1–n_bins
    return np.floor(percentiles * n_bins).astype(int) + 1

# Terapkan untuk tiap dimensi
df_lrfm['L_Score'] = score_from_rank(df_lrfm[lr_cols['L']])
df_lrfm['R_Score'] = score_from_rank(df_lrfm['R_inverted'])
df_lrfm['F_Score'] = score_from_rank(df_lrfm[lr_cols['F']])
df_lrfm['M_Score'] = score_from_rank(df_lrfm[lr_cols['M']])
```

```python
# 5. Gabungkan profil LRFM
df_lrfm['LRFM_Profile'] = (
    df_lrfm['L_Score'].astype(str) +
    df_lrfm['R_Score'].astype(str) +
    df_lrfm['F_Score'].astype(str) +
    df_lrfm['M_Score'].astype(str)
)
```

```python
# Asumsikan df_lrfm sudah berisi L_Score, R_Score, F_Score, M_Score

def assign_four_segment(row):
    L, R, F, M = row['L_Score'], row['R_Score'], row['F_Score'], row['M_Score']
    # 1. Loyal: skor tinggi di semua dimensi
    if L >= 4 and R >= 4 and F >= 4 and M >= 4:
        return 'Loyal'
    # 2. New: pelanggan baru, prioritaskan sebelum At Risk
    elif L <= 2:
        return 'New'
    # 3. At Risk: recency rendah (skor R kecil)
    elif R <= 2:
        return 'At Risk'
    # 4. Potential: sisanya (frekuensi & monetary relatif baik)
    else:
        return 'Potential'

# Terapkan ke DataFrame
df_lrfm['Four_Segment'] = df_lrfm.apply(assign_four_segment, axis=1)

# Verifikasi tidak ada kategori lain
print(df_lrfm['Four_Segment'].value_counts())
```

```python
df_lrfm['Four_Segment'].value_counts().plot(kind='bar', figsize=(6,4))
plt.title('Ukuran Segmen Sederhana')
plt.xlabel('Segment')
plt.ylabel('Jumlah Nasabah')
plt.tight_layout()
plt.show()
```

## Segmentasi Pelanggan Berdasarkan LRFM (Sederhana)

Kita membagi pelanggan ke dalam 4 segmen utama berdasarkan skor LRFM (Length, Recency, Frequency, Monetary):

| Segmen      | Definisi                                                                                                      | Kriteria Skor LRFM                                     |
|-------------|---------------------------------------------------------------------------------------------------------------|---------------------------------------------------------|
| **Loyal**     | Pelanggan yang paling berharga dan paling aktif: lama menjadi nasabah, transaksi baru-baru ini, sering, dan bernilai tinggi. | - **L_Score ≥ 4**  (Length: lama keanggotaan tinggi)  
               |                                                                                                               | - **R_Score ≥ 4**  (Recency: transaksi terakhir baru-baru ini)  
               |                                                                                                               | - **F_Score ≥ 4**  (Frequency: jumlah transaksi tinggi)  
               |                                                                                                               | - **M_Score ≥ 4**  (Monetary: nilai transaksi tinggi)        |
| **Potential** | Pelanggan dengan potensi tinggi: sering dan bernilai tinggi, meski keanggotaan atau recency bisa menengah.  | - **F_Score ≥ 4**  (Frequency tinggi)  
               |                                                                                                               | - **M_Score ≥ 4**  (Monetary tinggi)  
               |                                                                                                               | - **L_Score**, **R_Score**: tidak harus tinggi (boleh 2–5)     |
| **New**       | Pelanggan baru: keanggotaan masih pendek, belum memiliki riwayat transaksi panjang.                          | - **L_Score ≤ 2**  (Length: baru bergabung)  
               |                                                                                                               | - **R_Score**, **F_Score**, **M_Score**: bervariasi              |
| **At Risk**   | Pelanggan yang kemungkinan churn tinggi: jarang bertransaksi akhir‑akhir ini (recency rendah).               | - **R_Score ≤ 2**  (Recency: transaksi terakhir lama)  
               |                                                                                                               | - **L_Score**, **F_Score**, **M_Score**: bervariasi              |

### Penjelasan Masing‑Masing Segmen

1. **Loyal**  
   - Merupakan pelanggan paling berharga.  
   - Sudah lama menjadi nasabah, transaksi masih segar, sering bertransaksi, dan total nilai transaksi tinggi.  
   - Target: program penghargaan, upsell produk premium, referral.

2. **Potential**  
   - Pelanggan yang sering bertransaksi dan memiliki nilai transaksi tinggi, namun bisa jadi masih baru atau transaksi terakhirnya kurang segar.  
   - Target: campaign untuk meningkatkan frekuensi dan engagement (promo bundle, cashback).

3. **New**  
   - Pelanggan yang baru bergabung (Length rendah).  
   - Masih butuh edukasi dan pemahaman produk/layanan.  
   - Target: onboarding program, tutorial fitur, welcome offer.

4. **At Risk**  
   - Pelanggan yang lama tidak bertransaksi (Recency rendah).  
   - Berpotensi churn.  
   - Target: re‑engagement campaign (penawaran khusus, reminder manfaat layanan).
